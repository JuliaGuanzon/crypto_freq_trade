```python
parameters = ['asset_id', 
              'time', 
              'open',
              'close',
              'high',
              'low', 
              'volume', 
              'market_cap',
              'reddit_posts',
              'reddit_posts_score',
              'reddit_comments',
              'reddit_comments_score', 
              'tweets', 'tweet_spam',
              'tweet_followers',
              'tweet_quotes',
              'tweet_retweets', 
              'tweet_replies', 
              'tweet_favorites', 
              'tweet_sentiment1', 
              'tweet_sentiment2', 
              'tweet_sentiment3', 'tweet_sentiment4', 'tweet_sentiment5', 'tweet_sentiment_impact1', 'tweet_sentiment_impact2', 'tweet_sentiment_impact3', 'tweet_sentiment_impact4', 'tweet_sentiment_impact5', 'social_score', 'average_sentiment', 'sentiment_absolute', 'sentiment_relative', 'search_average', 'news', 'price_score', 'social_impact_score', 'correlation_rank', 'galaxy_score', 'volatility', 'alt_rank', 'alt_rank_30d', 'market_cap_rank', 'percent_change_24h_rank', 'volume_24h_rank', 'social_volume_24h_rank', 'social_score_24h_rank', 'medium', 'youtube', 'social_contributors', 'social_volume', 'price_btc']
```

```python
def cleandata(assets,assets_data):
    for item in assets["data"][0]["timeSeries"]:
        temp = {}
        for key,value in item.items():
            if key in parameters:
                temp[key]=value
        assets_data["timeSeries"].append(temp)
    return assets_data
```

```python
# def getdata():
#     url = f"https://api.lunarcrush.com/v2?data_points=100&change=1w&data=assets&symbol=BTC&key={api_key}"
#     response = requests.get(url)
#     assets = response.json()
    
#     return assets
start_date = 1605315600 
end_date = 1633827600

def getdata():
    url = f"https://api.lunarcrush.com/v2?data_points=700&interval=day&change=1d&data=assets&start={start_date}&end={end_date}&symbol=BTC&key={api_key}"
    response = requests.get(url)
    assets = response.json()
    
    return assets
```

```python
final_df_2020_2021 = pd.DataFrame(columns=parameters)
pd.set_option('display.max_columns', None)
```

```python
i=0
while i < 1:
    assets=getdata()
    assets_data = {
    
    "symbol": assets["data"][0]["symbol"],
    "price": assets["data"][0]["price"],
    "timeSeries":[]
    }    
    cleaned=cleandata(assets,assets_data)
    temp_df = pd.DataFrame(cleaned["timeSeries"])
    temp_df["time"]=pd.to_datetime(temp_df["time"],unit="s")

    final_df_2020_2021 = pd.concat([final_df_2020_2021,temp_df], axis=0,ignore_index=True)
   
    display(final_df_2020_2021)
    final_df_2020_2021.to_csv("final_df_2020_2021.csv")
    i += 1
    time.sleep(60)
    
    
```
